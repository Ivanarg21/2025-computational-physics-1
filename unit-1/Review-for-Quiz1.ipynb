## File permisions in Linux

- Every file and directory in Linux has associated permissions that control who can **read** (r), **write** (w), or **execute** (x) it.

- These permissions are assigned to three types of users:

 - the **user** (file owner, **u**),
 - the **group** (people working alongside the owner, **g**), and
 - **others** (everyone else, **o**).
 - all (user, group, and others, **a**)

- Each letter within a set represents a specific permission:

  - **`r` (read) (4):** Allows viewing the file's contents or listing a directory's contents.
  - **`w` (write) (2):** Allows modifying, saving, or deleting the file, or creating/deleting files within a directory.
  - **`x` (execute) (1):** Allows running a file as a program or entering a directory.
  - A hyphen (`-`) **(0)** in place of a letter means that permission is not granted.

### In order to copy and remove DIRECTORIES

**`cp -r`** "directory1" "directory2": Copy directory1 to directory2

**`rm -r "directory1"`**: Remove directory1

_____________________________________________________________________________________________________________

Low-level languages are close to the computer language. They need to be compiled into binary in order to be executed, which means that they are faster and more efficient, but they are difficult to write and read.
On the other hand, High-level languages are closer to the human language. An interpreter is neccesary in order to execute the script, which means that they are not as fast as Low-level languages, but they are easier to read and write, and they are also portable (can run in different kinds of computers with few or no modifications). Most people use High-level languages.

Some examples of each one:

High-level --> Python, Julia, MatLab, Mathematica.

Low-level --> Fortran, C.
